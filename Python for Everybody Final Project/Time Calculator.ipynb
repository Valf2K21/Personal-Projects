Task:
1. create a function that receives a list of strings, namely:
    - a start time in the 12-hour clock format that ends in either AM or PM
    - a duration time that indicates the number of hours and minutes
    - an optional starting day of the week, case-insensitive
2. the function should add the duration time to the start time and return a result
3. if the optional argument is not given, it should either return a (next day) or (n days later) prompt next to the result time
4. if the optional argument is given, it should return the result day of the week

Function Calls:
1. add_time("3:00 PM", "3:10")
2. add_time("11:30 AM", "2:32", "Monday")
3. add_time("11:43 AM", "00:20")
4. add_time("10:10 PM", "3:30")
5. add_time("11:43 PM", "24:20", "tueSday")
6. add_time("6:30 PM", "205.12")

```python
# import sys library to exit and send error messages, if needed
import sys

# user-defined function that will calculate user-given time
def add_time(startTime, durTime, optional):
    # initialize lists that will store start and duration hours and minutes
    hrs = []
    mins = []
    meridiem = ['AM', 'PM']
    days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

    # split user-given data
    amPM = startTime.split()
    startHrMin = amPM[0].split(':')
    durHrMin = durTime.split(':')

    # place newly-split data in their respective lists and variables
    merid = amPM[1]
    hrs.append(startHrMin[0])
    hrs.append(durHrMin[0])
    mins.append(startHrMin[1])
    mins.append(durHrMin[1])

    # send error message if user provided startTime and durTime minutes greater than 59
    if(int(durHrMin[1]) > 59 or int(durHrMin[1]) < 0):
        sys.exit('Error: Duration minutes must be between 0-59 only')

    if(int(startHrMin[1]) > 59 or int(startHrMin[1]) < 0):
        sys.exit('Error: Start minutes must be between 0-59 only')

    # send error message if user don't use 12-hour format
    if(int(hrs[0]) < 1 or int(hrs[0]) > 12):
        sys.exit('Error: Start time must be in 12-hour format only')

    # get minutes sum, divide by 60, and set remainder as finalMins, whole number quotient as addToHrs
    addMins = int(mins[0]) + int(mins[1])
    finalMins = addMins % 60 # to concatenate with results
    addToHrs = addMins // 60

    # if merid is PM, add 12 hours to represent AM in start hours
    if(merid == 'AM'):
        strHrs = int(hrs[0])

    elif(merid == 'PM'):
        strHrs = int(hrs[0]) + 12

    # send error message if user provided invalid meridiem
    else:
        sys.exit('Error: Meridiem must be AM or PM only')

    # get hours sum, divide by 12, and set remainder as finalHrs12, whole number quotient finalHrs to find finalMerid
    addHrs = strHrs + int(hrs[1]) + addToHrs
    finalHrs = addHrs // 12 
    finalHrs12 = addHrs % 12 # to concatenate with results

    # if-elif statement to get meridIndex
    if(merid == 'AM'):
        meridIndex = 0

    elif(merid == 'PM'):
        meridIndex = 1

    # use the sum of finalHrs and meridIndex to get finalMerid
    count = finalHrs + meridIndex
    counter = meridIndex

    # if addHrs don't go through a new day, use the next meridiem instead (e.g. AM to PM and vice-versa)
    if(addHrs < 24):
        if(addHrs < 12):
            finalMerid = 'AM'

        else:
            finalMerid = 'PM'


    # while-loop that will loop through AM and PM until it gets finalMerid
    while counter != count and addHrs >= 24:
        finalMerid = meridiem[meridIndex] # to concatenate with results
        if(meridIndex == 1):
            meridIndex = meridIndex - 1

        else:
            meridIndex = meridIndex + 1

        counter = counter + 1

    # while-loop to get days passed after adding start and dur
    daysPassed = 0
    if(addHrs >= 24):
        while True:
            difHrs = addHrs - 24
            divHrs = difHrs // 24

            if(divHrs != 0):
                daysPassed = daysPassed + 1

            elif(divHrs == 0):
                daysPassed = daysPassed + 1
                break

            addHrs = difHrs

    # get curDayIndex of given optional day
    if(optional != None):
        curDayIndex = days.index(optional.lower().capitalize())
        daysCount = 0

        # while-loop to loop through days list and return result day
        while daysCount != daysPassed + 1:
            resultDay = days[curDayIndex]

            if(curDayIndex == 6):
                curDayIndex = 0

            else:
                curDayIndex = curDayIndex + 1

            daysCount = daysCount + 1

    # if finalHrs12 is 0, convert it to 12
    if(finalHrs12 == 0):
        finalHrs12 = 12

    # if finalMins is less than 9, append a 0 before it
    if(finalMins < 10):
        finalMins = '0' + str(finalMins)

    # if-else statement that will print results depending on user-provided data
    if(optional == None):
        finalDayA = '(' + str(daysPassed) + ' days later)'

        if(daysPassed == 0):
            finalDayA = ''

        elif(daysPassed == 1):
            finalDayA = '(next day)'
        print(str(finalHrs12) + ':' + str(finalMins), finalMerid, finalDayA)

    else:
        finalDayA = '(' + str(daysPassed) + ' days later)'
        
        if(daysPassed == 0):
            finalDayA = ''

        elif(daysPassed == 1):
            finalDayA = '(next day)'
        finalDayB = resultDay
        print(str(finalHrs12) + ':' + str(finalMins), finalMerid + ',', finalDayB, finalDayA)
```

```python
# while-loop to keep the program running and ask users for time information
while True:
    startTime = input('Type the start time (12-hour format only) (ex: 12:00 PM): ')
    durTime = input('Type the duration time (hours:minutes format) (ex: 204:59): ')
    optional = input('Type the day of start time (optional) (Monday to Sunday only): ')
    contChoice = input('Calculate another time? (Yes/No): ')

    # set default optional value if user don't provide any
    if(optional == ''):
        optional = None

    # call function and supply arguments for it to use, then either continue or end program depending on user
    if(contChoice.lower() == 'yes'):
        add_time(startTime.upper(), durTime, optional)
        continue

    else:
        add_time(startTime.upper(), durTime, optional)
        break

    # clear variables contents for new iteration
    startTime = ''
    durTime = ''
    optional = ''
```
