Task:
1. Create a function that receives a list of strings that are arithmetic problems
2. Return the problems arranged vertically and side-by-side
3. The function should optionally take a second argument
4. If the second argument is True, display the answers

Function Calls:
1. arithmetic_arranger(["32 + 698", "3801 - 2", "45 + 43", "123 + 49"])
2. arithmetic_arranger(["32 + 8", "1 - 3801", "9999 + 9999", "523 - 49"], True)

```python
# import sys library to exit and send error messages, if needed
import sys

# user-defined function that will arrange the problems vertically and solve it
def arithmetic_arranger(problems, ansChoice):
    # initialize lists that will store operands, symbols, and answers
    operA = []
    symbols = []
    operB = []
    answers = []

    # initialize variables containing placeholder positions for conversion
    operALine = ' s1num1     s2num2     s3num3     s4num4     s5num5'
    operBLine = '-1s1num1    -2s2num2    -3s3num3    -4s4num4    -5s5num5'
    dashLine = 'dash1    dash2    dash3    dash4    dash5'
    answerLine = 's1ans1    s2ans2    s3ans3    s4ans4    s5ans5'

    # initialize counters and starting placeholders to change from
    posOldCount = 0
    posNewCount = None
    negOldCount = 0
    negNewCount = None
    operPlace = 'num1'
    dashPlace = 'dash1'
    ansPlace = 'ans1'
    spacePlace = 's1'

    # for-loop to split each arithmetic problem and group their parts in their respective variables
    countP = 0
    for prob in problems:
        finalP = countP + 1
        probPart = prob.split()
        operA.append(probPart[0])
        symbols.append(probPart[1])
        operB.append(probPart[2])
        countP = finalP
    
    # send error message if the program received more than 5 arithmetic problems
    if(countP > 5):
        sys.exit('Error: Too many problems.')

    # for-loop to do multiple tasks commented below
    for symb in symbols:
        # count the number of iterations
        posNewCount = posOldCount + 1
        negNewCount = negOldCount - 1

        # save item value of each operand to separate variables for the next loops to use
        holderOperA = operA[posNewCount - 1].split()
        holderOperB = operB[posNewCount - 1].split()

        # for-loop to count the digits of each temporarily stored operands
        index = 0
        for opA in holderOperA:
            countA = 0

            # for-loop to count the number of digits in each operA operand temporarily stored in opA
            for a in opA:
                countA = countA + 1

            # while-loop that will run in a [1:1] operA:operB ratio
            while True:
                opB = holderOperB[index]
                countB = 0

                # for-loop to count the number of digits in each operB operand temporarily stored in opB
                for b in opB:
                    countB = countB + 1

                newindex = index + 1
                index = newindex
                break

        # send error message if the program received an operand with illegal character in it
        try:
            convertA = int(opA)
            convertB = int(opB)
        
        except:
            convertA = -1
            convertB = -1
        
        if(convertA == -1 or convertB == -1):
            sys.exit('Error: Numbers must only contain digits.')

        # send error message if one of the operands to add has more than four digits
        if(countA > 4 or countB > 4):
            sys.exit('Error: Numbers cannot be more than four digits.')
        
        
        # if-elif statements to answer arithmetic problems and append it to answers list
        if(symb == '+'):
            sumAns = int(opA) + int(opB)
            answers.append(str(sumAns))

        elif(symb == '-'):
            difAns = int(opA) - int(opB)
            answers.append(str(difAns))

        # send error message if the program received an illegal operator
        else:  
            sys.exit("Error: Operator must be '+' or '-'.")

        # set operand, dash, answer, and space placeholders to change
        operChange = operPlace.replace(str(posOldCount), str(posNewCount))
        dashChange = dashPlace.replace(str(posOldCount), str(posNewCount))
        ansChange = ansPlace.replace(str(posOldCount), str(posNewCount))
        spaceChange = spacePlace.replace(str(posOldCount), str(posNewCount))

        # change symbol placeholder to real symbol and save it before inserting operB
        placedSymb = operBLine.replace(str(negNewCount), symb)
        operBLine = placedSymb

        # change operA and operB operand placeholders, and answers placeholder  to item values
        placedOperA = operALine.replace(operChange, operA[posNewCount - 1])
        placedOperB = operBLine.replace(operChange, operB[posNewCount - 1])
        placedAnswers = answerLine.replace(ansChange, answers[posNewCount - 1])

        # update operALine, operBLine, and answerLine placeholder positions before adding spaces
        operALine = placedOperA
        operBLine = placedOperB
        answerLine = placedAnswers

        # for-loop to count the digits of temporarily stored answer
        countC = 0
        for ans in answers[posNewCount - 1]:
            countC = countC + 1

        # if opA has more digits than opB, replace spaceChange placeholder with specific number of spaces to operBLine
        # also, use a specific number of dashes depending on the number of digits of opA
        if(countA > countB):
            if(countA == 4):
                placedDash = dashLine.replace(dashChange, '------')
                if(countB == 3):
                    placedOperA = operALine.replace(spaceChange, ' ')
                    placedOperB = operBLine.replace(spaceChange, '  ')

                elif(countB == 2):
                    placedOperA = operALine.replace(spaceChange, ' ')
                    placedOperB = operBLine.replace(spaceChange, '   ')

                elif(countB == 1):
                    placedOperA = operALine.replace(spaceChange, ' ')
                    placedOperB = operBLine.replace(spaceChange, '    ')

            elif(countA == 3):
                placedDash = dashLine.replace(dashChange, '-----')
                if(countB == 2):
                    placedOperA = operALine.replace(spaceChange, ' ')
                    placedOperB = operBLine.replace(spaceChange, '  ')

                elif(countB == 1):
                    placedOperA = operALine.replace(spaceChange, ' ')
                    placedOperB = operBLine.replace(spaceChange, '   ')

            elif(countA == 2):
                placedDash = dashLine.replace(dashChange, '----')
                if(countB == 1):
                    placedOperA = operALine.replace(spaceChange, ' ')
                    placedOperB = operBLine.replace(spaceChange, '  ')

        # if opB has more digits than opA, replace spaceChange placeholder with specific number of spaces to operALine
        # also, use a specific number of dashes depending on the number of digits of opB
        elif(countA < countB):
            if(countB == 4):
                placedDash = dashLine.replace(dashChange, '------')
                if(countA == 3):
                    placedOperA = operALine.replace(spaceChange, '  ')
                    placedOperB = operBLine.replace(spaceChange, ' ')

                elif(countA == 2):
                    placedOperA = operALine.replace(spaceChange, '   ')
                    placedOperB = operBLine.replace(spaceChange, ' ')

                elif(countA == 1):
                    placedOperA = operALine.replace(spaceChange, '    ')
                    placedOperB = operBLine.replace(spaceChange, ' ')

            elif(countB == 3):
                placedDash = dashLine.replace(dashChange, '-----')
                if(countA == 2):
                    placedOperA = operALine.replace(spaceChange, '  ')
                    placedOperB = operBLine.replace(spaceChange, ' ')

                elif(countA == 1):
                    placedOperA = operALine.replace(spaceChange, '   ')
                    placedOperB = operBLine.replace(spaceChange, ' ')
    
            elif(countB == 2):
                placedDash = dashLine.replace(dashChange, '----')
                if(countA == 1):
                    placedOperA = operALine.replace(spaceChange, '  ')
                    placedOperB = operBLine.replace(spaceChange, ' ')

        # if opA and opB have the same amount of digits, use a specific number of dashes for them
        elif(countA == countB):
            if(countA == 4):
                placedDash = dashLine.replace(dashChange, '------')
                placedOperA = operALine.replace(spaceChange, ' ')
                placedOperB = operBLine.replace(spaceChange, ' ')

            elif(countA == 3):
                placedDash = dashLine.replace(dashChange, '-----')
                placedOperA = operALine.replace(spaceChange, ' ')
                placedOperB = operBLine.replace(spaceChange, ' ')

            elif(countA == 2):
                placedDash = dashLine.replace(dashChange, '----')
                placedOperA = operALine.replace(spaceChange, ' ')
                placedOperB = operBLine.replace(spaceChange, ' ')

            elif(countA == 1):
                placedDash = dashLine.replace(dashChange, '---')
                placedOperA = operALine.replace(spaceChange, ' ')
                placedOperB = operBLine.replace(spaceChange, ' ')

        # add spaces to newly-placed answers
        if(countA == 4 or countB == 4):
            if(countC == 5):
                placedAnswers = answerLine.replace(spaceChange, ' ')

            elif(countC == 4):
                placedAnswers = answerLine.replace(spaceChange, '  ')

            elif(countC == 3):
                placedAnswers = answerLine.replace(spaceChange, '   ')

            elif(countC == 2):
                placedAnswers = answerLine.replace(spaceChange, '    ')

            elif(countC == 1):
                placedAnswers = answerLine.replace(spaceChange, '     ')

        elif(countA == 3 or countB == 3):
            if(countC == 4):
                placedAnswers = answerLine.replace(spaceChange, ' ')

            elif(countC == 3):
                placedAnswers = answerLine.replace(spaceChange, '  ')

            elif(countC == 2):
                placedAnswers = answerLine.replace(spaceChange, '   ')

            elif(countC == 1):
                placedAnswers = answerLine.replace(spaceChange, '    ')

        elif(countA == 2 or countB == 2):
            if(countC == 3):
                placedAnswers = answerLine.replace(spaceChange, ' ')

            elif(countC == 2):
                placedAnswers = answerLine.replace(spaceChange, '  ')

            elif(countC == 1):
                placedAnswers = answerLine.replace(spaceChange, '   ')

        elif(countA == 1 or countB == 1):
            if(countC == 2):
                placedAnswers = answerLine.replace(spaceChange, ' ')

            elif(countC == 1):
                placedAnswers = answerLine.replace(spaceChange, '  ')

        # set new values as old values after using it for the next loop iteration to use
        posOldCount = posNewCount
        negOldCount = negNewCount
        operPlace = operChange
        dashPlace = dashChange
        ansPlace = ansChange
        spacePlace = spaceChange
        operALine = placedOperA
        operBLine = placedOperB
        dashLine = placedDash
        answerLine = placedAnswers

    # while-loop to remove remaining placeholders
    while posOldCount != 5 and negOldCount != -5:
        posNewCount = posOldCount + 1
        negNewCount = negOldCount - 1

        # set remaining operand, dash, and space placeholders to remove
        operChange = operPlace.replace(str(posOldCount), str(posNewCount))
        dashChange = dashPlace.replace(str(posOldCount), str(posNewCount))
        spaceChange = spacePlace.replace(str(posOldCount), str(posNewCount))

        # remove remaining operand and space placeholders in operALine
        placedOperA = operALine.replace(operChange, '')
        operALine = placedOperA.replace(spaceChange, '')

        # remove remaining operand, space, and operator placeholders in operBLine
        placedOperB = operBLine.replace(operChange, '')
        operBLine = placedOperB.replace(spaceChange, '')
        placedOperB = operBLine.replace(str(negNewCount), '')

        # remove remaining dash placeholders in dashLine
        placedDash = dashLine.replace(dashChange, '')

        # remove remaining space and answer placeholders in answerLine
        placedAnswers = answerLine.replace(spaceChange, '')
        answerLine = placedAnswers.replace('an', '')

        operBLine = placedOperB
        dashLine = placedDash

        # set new values as old values after using it for the next loop iteration to use
        operPlace = operChange
        dashPlace = dashChange
        spacePlace = spaceChange
        posOldCount = posNewCount
        negOldCount = negNewCount

    # return results by printing them when function is called
    print(operALine.rstrip())
    print(operBLine.rstrip())
    print(dashLine.rstrip())
    
    if(ansChoice == True):
        print(answerLine.rstrip())
```

```python
# initialize list that will store user-provided arithmetic problems
problems = []

# while-loop to keep the program running and ask users for arithmetic problems
while True:
    problem = input('Type a math problem: ')
    contChoice = input('Add another one? (Yes/No): ')
    problems.append(problem)

    if(contChoice.lower() == 'yes'):
        continue

    # ask users if they need answers or not
    else:
        ansChoice = input('Show answers? (Yes/No): ')

        if(ansChoice.lower() == 'yes'):
            ansChoice = True

        else:
            ansChoice = False

    break

# call function and supply arguments for it to use
arithmetic_arranger(problems, ansChoice)
```
