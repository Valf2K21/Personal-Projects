```python
# a hat class with a draw method
class Hat:
    # use init function to make variables that objects can use independently
    def __init__(self, **kwargs):
        # initialize contents and totalballs variables for later use
        self.contents = []
        self.totalBalls = 0
        self.ballsColors = []

        # for-loop to loop through each key-value pairs of user-given arguments
        for key, value in kwargs.items():
            # add current value to finally get self.totalBalls
            self.totalBalls = self.totalBalls + value

            # store available colors of balls to list and use it as experiment checker later
            self.ballsColors.append(key)

            # convert key-value balls to string balls and append to self.contents
            ctr = 0
            while(ctr < value):
                self.contents.append(key)
                ctr = ctr + 1

    # a draw method that will draw user-defined number of balls
    def draw(self, drawBalls):
        # initialize pickedBalls for picker use below
        pickedBalls = []

        # don't continue process if drawBalls is greater than totalBalls
        if(drawBalls > self.totalBalls):
            return None

        ctr = 0
        # make a copy of self.contents and use it instead of the original self.contents
        contentsCopy = copy.copy(self.contents)

        # while-loop to draw balls, append, then remove in list so they won't be picked again
        while(ctr < drawBalls):
            choice = random.choice(contentsCopy)
            pickedBalls.append(choice)
            contentsCopy.remove(choice)
            ctr = ctr + 1

        # for-loop to convert pickedBalls strings list into key-word dictionary
        ballsDict = {}
        ballsDictColors = []
        for i in pickedBalls:
            # if current ball color not in ballsDict, make a new key-value pair out of it
            if(i not in ballsDict):
                ballsDict[str(i)] = 1

            # else, just add 1 to existing key-value pair of that color
            else:
                ballsDict[str(i)] = ballsDict[str(i)] + 1

            if(i not in ballsDictColors):
                ballsDictColors.append(i)

        # return value for experiment function to use
        print(ballsDict)
        return ballsDict, ballsDictColors

    # a returnColors method that will return list of balls colors in current hat
    def returnColors(self):
        # initialize ballsColors for experiment checker to use
        ballsColors = self.ballsColors

        # return value for experiment function to use
        return ballsColors

    # test method
    def get(self):
        print(self.contents, self.totalBalls)

# an experiment method that will commence experiments repeatedly until it meets user-defined number of experiments to execute
def experiment(hat, expected_balls, num_balls_drawn, num_experiments):
    # initialize ballsColors taken from hat class' returnColors method
    ballsColors = hat.returnColors()

    # for-loop to check if any expected_balls color is not in ballsColors list
    for i in expected_balls:
        if(i not in ballsColors):
            print('Error: One or more expected colors is/are not in the hat. Experiment stopped.')
            return None

    experimentCtr = 0
    correctCtr = 0

    # while-loop to commence experiments repeatedly
    while(experimentCtr < num_experiments):
        # start random picking experiment
        ballsDict, ballsDictColors = hat.draw(num_balls_drawn)

        # initialize expectedStatus empty list for ball checking later
        expectedStatus = []

        # for-loop to go through each expected_balls color
        for e in expected_balls:
            # before checking, check if one or more expected colors are not in picked ball list; if yes, skip to next ball list
            if e not in ballsDictColors:
                print('Error: One or more expected colors is/are not in the picked balls list. Skipping to next experiment.')
                expectedStatus = []
                break

            # for-loop to go through each ballsDict color
            for b in ballsDict:
                # if expected color e and ballsDict color b is the same, compare their values
                if(e == b):
                    # if current color's expected value is less or equal to its ballsDict value, append 'pass' to expectedstatus list
                    if(expected_balls[e] <= ballsDict[b]):
                        expectedStatus.append('pass')
                        break

                    # else, append 'fail'
                    else:
                        expectedStatus.append('fail')
                        break

                # else, continue to next ballsDict color and compare again
                else:
                    continue

        # if expectedstatus list is not empty and there's no 'fail' in it, add 1 to correctctr counter to be used for probability solving
        if(expectedStatus != [] and 'fail' not in expectedStatus):
            correctCtr = correctCtr + 1

        # add 1 to experimentctr for while-loop's use and also for usage on probability solving
        experimentCtr = experimentCtr + 1

    # calculate probability and print it at the end of comparing all ballslist to expecte_balls
    probSol = correctCtr / experimentCtr
    print('The probability of finding', expected_balls, ' on', experimentCtr, ' experiments is:', f'{probSol:.2%}')

    # test: print values to check if program works
    print('Experiment Counter:', experimentCtr, '\nCorrect Counter:', correctCtr)
```
