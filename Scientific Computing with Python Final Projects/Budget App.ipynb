    - a deposit method that accepts an amount and description and appends it to the ledger list in the form of {'amount' : amount, 'description' : description}; if no description is given, default it to an empty string
    - a withdraw method that is similar to the deposit method and will either return True or False whether the withdrawal took place or not, but the amount passed in should be stored in the ledger as a negative number; if not enough funds, nothing should be added to the ledger
    - a get_balance method that returns the current balance of the budget category based on the deposits and withdrawals that have occurred
    - a transfer method that accepts an amount and another budget category as arguments, which will add a withdraw amount and the description 'Transfer to [destination budget category]', while it also adds a deposit amount and the description 'Transfer from [source budget category]'; if not enough funds, nothing should be added to either ledgers and will return either True or False whether the transfer took place or not
    - a check_funds method that accepts an amount as an argument, which will return False if the amount is greater than the balance of the budget category and returns True otherwise; this method should be used by both the withdraw and transfer methods

```python
# a category class that contains five methods or actions that can be executed
class Category:
    # variables that will store default category name, amount, description, and transfer category name values
    oldCat = ''
    newCat = ''

    # use init function to make a ledger that objects can use independently
    def __init__(self, oldCat):
        self.oldCat = oldCat

        # set default values of ledger, balance, and amntSmaller
        self.ledger = []
        self.balance = 0
        self.amntSmaller = None

    # a deposit method that accepts a positive amount and description
    def deposit(self, amnt, desc):
        # append amount and description to ledger list
        addToLedger = {desc:amnt}
        self.ledger.append(addToLedger)

        # update self.balance by adding new amount in it
        self.balance = self.balance + float(amnt)

        # print receipt every transaction
        if(self.oldCat.lower() == 'food'):
            print('\n*************Food*************')

        elif(self.oldCat.lower() == 'clothing'):
            print('\n***********Clothing***********')

        elif(self.oldCat.lower() == 'auto'):
            print('\n*************Auto*************')

        elif(self.oldCat.lower() == 'entertainment'):
            print('\n********Entertainment*********')

        for item in self.ledger:
            for i in item:
                print(f'{i[0:23] : <23}{str(item[i])[0:7] : >7}')

        editedBalance = format(self.balance, '.2f')
        print('Total: ' + str(editedBalance) + '\n')

    # a withdraw method that accepts a negative amount and description
    def withdraw(self, amnt, desc):
        # call check_funds method to check funds
        self.check_funds(amnt)

        # if self.amntSmaller is True, allow withdraw action
        if(self.amntSmaller == True):
            # append amount and description to ledger list
            addToLedger = {desc:'-' + amnt}
            self.ledger.append(addToLedger)

            # update self.balance by adding new amount in it
            self.balance = self.balance - float(amnt)

            # print receipt every transaction
            if(self.oldCat.lower() == 'food'):
                print('\n*************Food*************')

            elif(self.oldCat.lower() == 'clothing'):
                print('\n***********Clothing***********')

            elif(self.oldCat.lower() == 'auto'):
                print('\n*************Auto*************')

            elif(self.oldCat.lower() == 'entertainment'):
                print('\n********Entertainment*********')

            for item in self.ledger:
                for i in item:
                    print(f'{i[0:23] : <23}{str(item[i])[0:7] : >7}')

            editedBalance = format(self.balance, '.2f')
            print('Total: ' + str(editedBalance) + '\n')

    # a get_balance method
    def get_balance(self):
        # print current balance
        editedBalance = format(self.balance, '.2f')
        print('Balance: ' + str(editedBalance))

    # a transfer method that accepts amount and another separate category
    def transfer(self, amnt, newCat):
        # call check_funds method to check funds
        self.check_funds(amnt)

        # if self.amntSmaller is True, allow transfer action
        if(self.amntSmaller == True):
            # call withdraw method of oldCat to deduct funds to transfer
            desc = 'Transfer to ' + newCat.capitalize()
            self.withdraw(amnt, desc)

            # call deposit method of newCat to add funds to transfer
            desc = 'Transfer from ' + self.oldCat
            if(newCat == 'food'):
                f.deposit(amnt, desc)

            elif(newCat == 'clothing'):
                c.deposit(amnt, desc)

            elif(newCat == 'auto'):
                a.deposit(amnt, desc)

            elif(newCat == 'entertainment'):
                e.deposit(amnt, desc)

    # a check_funds method
    def check_funds(self, amnt):
        # compare new amount to self.balance and set self.amntSmaller value
        if(float(amnt) > self.balance):
            self.amntSmaller = False

        else:
            self.amntSmaller = True
```

```python
# import sys and itertools libraries to exit and send error messages and create bar chart
import sys
import itertools

# construct object categories and store them in their respective variables (which will convert into 'self' variable in class category methods)
f = Category('Food')
c = Category('Clothing')
a = Category('Auto')
e = Category('Entertainment')

# while-loop to keep the program running
while True:
    # print program header and ask the user for instructions
    print('Valf2K21 Budget App')
    chosenOper = input('Enter operation to do (deposit/withdraw/balance/transfer/exit): ')

    # exit program if user says so
    if(chosenOper == 'exit'):
        break

    # ask user for category to modify
    chosenCat = input('Enter category to modify ( [F] food / [C] clothing / [A] auto / [E] entertainment ): ')

    # tell the chosen category object to run a specific method depending on the user's chosen operation
    if(chosenOper.lower() == 'deposit'):
        # ask users for data required by this method
        descrip = input('Enter deposit description (max 23 characters only): ')
        amount = float(input('Enter deposit amount (max of 9999.99 only): '))
        editedAmount = format(amount, '.2f')

        if(chosenCat == 'food'):
            f.deposit(editedAmount, descrip)

        elif(chosenCat == 'clothing'):
            c.deposit(editedAmount, descrip)

        elif(chosenCat == 'auto'):
            a.deposit(editedAmount, descrip)

        elif(chosenCat == 'entertainment'):
            e.deposit(editedAmount, descrip)

    elif(chosenOper.lower() == 'withdraw'):
        # ask users for data required by this method
        descrip = input('Enter withdrawal description (max 23 characters only): ')
        amount = float(input('Enter withdrawal amount (max of 9999.99 only): '))
        editedAmount = format(amount, '.2f')

        if(chosenCat == 'food'):
            f.withdraw(editedAmount, descrip)

        elif(chosenCat == 'clothing'):
            c.withdraw(editedAmount, descrip)

        elif(chosenCat == 'auto'):
            a.withdraw(editedAmount, descrip)

        elif(chosenCat == 'entertainment'):
            e.withdraw(editedAmount, descrip)

    elif(chosenOper.lower() == 'balance'):
        if(chosenCat == 'food'):
            f.get_balance()

        elif(chosenCat == 'clothing'):
            c.get_balance()

        elif(chosenCat == 'auto'):
            a.get_balance()

        elif(chosenCat == 'entertainment'):
            e.get_balance()

    elif(chosenOper.lower() == 'transfer'):
        # ask users for data required by this method
        amount = float(input('Enter transfer amount (max of 9999.99 only): '))
        categ = input('Enter category to transfer amount to: ')
        editedAmount = format(amount, '.2f')

        # send error message if chosenCat and categ is the same
        if(categ == chosenCat):
            # change to proper error message print('error: cannot transfer to same category')
            break

        if(chosenCat == 'food'):
            f.transfer(editedAmount, categ)

        elif(chosenCat == 'clothing'):
            c.transfer(editedAmount, categ)

        elif(chosenCat == 'auto'):
            a.transfer(editedAmount, categ)

        elif(chosenCat == 'entertainment'):
            e.transfer(editedAmount, categ)
```

```python
# a create_spend_chart function to create percentage spent chart
def create_spend_chart(barCatList):
    # for-loop through each user-input category listed in barCatList
    toAdd = []
    lowerline = '    -'
    for item in barCatList:
        lowerline = lowerline + '---'
        # if-elif statements to check the current category temporarily stored in item
        if(item == 'Food'):
            # initialize foodTotal and calculate withdrawal amounts stored in food's own ledger
            foodTotal = 0
            for thing in f.ledger:
                for t in thing:
                    if(float(thing[t]) < 0):
                        foodTotal = foodTotal + float(thing[t])
            foodAppend = {'foodTotal' : foodTotal}
            toAdd.append(foodAppend)

        elif(item == 'Clothing'):
            # initialize clothingTotal and calculate withdrawal amounts stored in clothing's own ledger
            clothingTotal = 0
            for thing in c.ledger:
                for t in thing:
                    if(float(thing[t]) < 0):
                        clothingTotal = clothingTotal + float(thing[t])
            clothingAppend = {'clothingTotal' : clothingTotal}
            toAdd.append(clothingAppend)

        elif(item == 'Auto'):
            # initialize autoTotal and calculate withdrawal amounts stored in auto's own ledger
            autoTotal = 0
            for thing in a.ledger:
                for t in thing:
                    if(float(thing[t]) < 0):
                        autoTotal = autoTotal + float(thing[t])
            autoAppend = {'autoTotal' : autoTotal}
            toAdd.append(autoAppend)

        elif(item == 'Entertainment'):
            # initialize entertainmentTotal and calculate withdrawal amounts stored in entertainment's own ledger
            entertainmentTotal = 0
            for thing in e.ledger:
                for t in thing:
                    if(float(thing[t]) < 0):
                        entertainmentTotal = entertainmentTotal + float(thing[t])
            entertainmentAppend = {'entertainmentTotal' : entertainmentTotal}
            toAdd.append(entertainmentAppend)

    finalTotal = 0
    percentages = []
    for item in toAdd:
        # for-loop to add all totals to get finalTotal to divide to get percentage
        for i in item:
            finalTotal = finalTotal + item[i]

    for item in toAdd:
        # for-loop to calculate all percentages of withdrawal of each category when divided to total withdrawals, then add to see if it is 100%
        for i in item:
            if(i == 'foodTotal'):
                percentFood = (item[i] / finalTotal) * 100
                percentages.append(percentFood)

            elif(i == 'clothingTotal'):
                percentClothing = (item[i] / finalTotal) * 100
                percentages.append(percentClothing)

            elif(i == 'autoTotal'):
                percentAuto = (item[i] / finalTotal) * 100
                percentages.append(percentAuto)

            elif(i == 'entertainmentTotal'):
                percentEntertainment = (item[i] / finalTotal) * 100
                percentages.append(percentEntertainment)

    # print layout of barchart (not complete yet)
    n = 100
    letterO = ''
    print('Percentage spent by category')

    # while-loop to lopp through each bar lines in accordance to the y-axis n
    while n >= 0:
        # for-loop to add letter o to visualize bars in bar chart
        ctr = 0
        for item in barCatList:
            if(item == 'Food' or item == 'Clothing' or item == 'Auto' or item == 'Entertainment'):
                if(round(percentages[ctr], -1) >= n):
                    letterO = letterO + 'o  '

                else:
                    letterO = letterO + '   '
            ctr = ctr + 1

        # print bar line after adding o's then go to the next line and reset it
        print(f'{n : >3}' + '| ' + letterO)
        n = n - 10
        letterO = ''

    # print lowerline and categories in x-axis    
    print(lowerline)

    # print x-axis categories in vertical position
    for i in itertools.zip_longest(*barCatList, fillvalue=" "):
        if any(j != ' ' for j in i):
            print('     ' + '  '.join(i))
```
